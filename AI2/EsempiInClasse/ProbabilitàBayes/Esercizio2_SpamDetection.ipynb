```python
def teorema_bayes(p_a_dato_b, p_b, p_a):
    return (p_a_dato_b * p_b) / p_a

p_Spam = 0.20 #P(Spam)
p_Keyword_dato_Spam = 0.90 #P(Keyword|Spam)
p_Keyword_dato_Non_Spam = 0.60 #P(Keyword|Non-Spam)

p_Keyword = p_Keyword_dato_Spam * p_Spam + p_Keyword_dato_Non_Spam * (1 - p_Spam)

#calcola P(Spam|Keyword)
p_Spam_dato_Keyword = teorema_bayes(
    p_a_dato_b = p_Keyword_dato_Spam,
    p_b = p_Spam,
    p_a = p_Keyword
)

print(p_Spam_dato_Keyword)
```
